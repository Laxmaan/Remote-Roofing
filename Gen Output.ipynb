```python
import json
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
with open('bboxes.json','r') as f:
    data = json.load(f)
```

```python
keys = list(data.keys())
```

```python
i=0
img = plt.imread(keys[i],-1)
plt.imshow(img)
img.shape
```

```python
bboxes = data[keys[i]]
```

```python
bboxes = list(map(lambda x: list(map(lambda y: float(y), x.strip().split(','))),bboxes))
```

```python
for x,y,w,h in bboxes:
    ymin,xmin,ymax,xmax = int(y),int(x),int(y+h),int(x+w)
    img = cv2.rectangle(img,(xmin,ymin),(xmax,ymax),(0,255,0,255),2)
```

```python
plt.imshow(img)
```
